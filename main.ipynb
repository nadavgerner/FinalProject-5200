```python
import pandas as pd
```

```python
traffic = pd.read_csv("data/dc_traffic_volume_2023.csv")
traffic["FROMDATE"] = pd.to_datetime(traffic["FROMDATE"], unit="ms", errors="coerce")
```

```python
traffic = traffic.sort_values('FROMDATE')
traffic.head()
```

```python
crashes = pd.read_csv("data/dc_crashes.csv")
date_cols = ["REPORTDATE", "FROMDATE", "TODATE", "LASTUPDATEDATE"]
for col in date_cols:
    if col in crashes.columns:
        crashes[col] = pd.to_datetime(crashes[col], unit="ms", errors="coerce")
```

```python
crashes = crashes.sort_values('REPORTDATE')
crashes.head()
```
